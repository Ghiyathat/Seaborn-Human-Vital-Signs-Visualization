Human Vital Signs Visualization with Seaborn

```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading dataset
vitalsigns_df = pd.read_csv('./human_vital_signs_dataset_2024_2000_data.csv')
vitalsigns_df
```

```python
# Checking the dataset
vitalsigns_df.head(10)
```

DISTRIBUTION OF PATIENTS AGE

```python
# Distribution of patients age

sns.histplot(data= vitalsigns_df, x= "Age", bins= 20, kde= True, color= "green")
plt.title("Distribution of Patient Age")
plt.xlabel("Age (years)")
plt.ylabel("Count")
plt.show()
```

EXPLANATION
1. This histogram shows the distribution of patient's age showing the frequency of patients within various age groups
2. The x- axis represents the age ranges of the patients starting from around 20 years up to 90 years, with bins representing 5- year intervals
3. The y- axis indicates the number of patients falling within each age bin
4. The distribution is somewhat irregular, showing multiple peaks indicating that patient number are not uniformly distributed across all age groups
5. There are notable high patient count around the age 30- 35, 55- 60' and 65-70, and a slight one for 80- 85 years. And there are lower counts in age range for early 20s, mid-40s and early 70s
6. The histogram reveals skewness, clusters and potential age-based risk groups. It identify whether patient sample skews younger or older.

BOXPLOTS FOR COMPARING GROUPS
HEART RATE DISTRIBUTION BY GENDER

```python
# Boxplots for comparing groups
# Heart Rate Distribution by Gender
sns.boxplot(data= vitalsigns_df, x="Gender", y="Heart Rate", hue="Gender", palette=["pink","skyblue"])
plt.title("Heart Rate Distribution by Gender")
plt.show()
```

EXPLANATION
1. The boxplot shows the heart rate distribution by gender
2. It shows that the median heart rate is the line inside each box, here for female; it is approximately 79 bpm while male is about the same thing , indicating a similar central tendency
3. It also shows the interquartile range(IQR), the box itself represents the IQR which contains the middle 50% of the data. For males, it ranges from around 70 bpm to 89 bpm, while females ranges from 69 bpm to 89 bpm. This indicates that a similar spread in the central 50% of heart rates for both genders
4. The whiskers, which is the line extending from the box indicate the range of data excluding outliers
5. For both genders, the whisker extend close to 60 bpm at the lower end and 99 bpm at the upper end. This suggest a similar overall range of heart rates
6. All data points fall within the calculated range. 
In conclusion, the heart rate distributors for male and female appear to be very similar in terms of median, spread and overall range, with no apparent outliers in either group because it shows median and quartiles by comparing clinical variables across patient demographies

PATIENT DEMOGRAPHICS
VIOLIN PLOT FOR SHAPING DISTRIBUTION
BODY TEMPERATURE DISTRIBUTION BY RISK CATEGORY

```python
# Body Temperature Distribution by risk category
sns.violinplot(data= vitalsigns_df, x= "Risk Category", y= "Body Temperature", hue= "Risk Category", palette=["red","skyblue"])
plt.title("Body Temperature Distribution by Risk Category")
plt.show()
```

EXPLANATION
1. The violin plot shows the distribution of body temperatures for the two main risk categories - High and Low
2. Both categories show a concentation of body temperatures around the normal range slightly between 36.5 and 37 degree celsius
3. The high risk represented in red violin is slightly wider and more spread compared to the low risk category especially towards the lower temperature, indicating a relatively high density of individuals with body temperatures around 36.75 and another around 36.35 degree celcius
4. The median is around 36.7 degree celsius, the interquartile range with the thick gray bar and the overall range with the thin gray line are visually wider showing more variability in body temperature for the high-risk group
5. The low risk in blue shows that the distribution is narrower and more concentrated around the central body temperture suggesting less variability
6. The peak density is around 36.75 degree celsius, the median is around 36.7 degree celsius as the interquartile range and overall range are visually narrower than the high-risk group
7. In conclusion, it shows that the high risk category shows a broader range and higher variability in body temperatures, while the low risk indicates a more tightly clustered distribution around the median. The violin plots combines boxplot with kernel density showing both spread and shape, indicating at high risk patients have more variable in body temperature.


```python
HEATMAP FOR CORRELATIONS AND PATTERNS
CORRELATION HEATMAP OF CLINICAL VARIABLES
```

```python
# Correlation heatmap of clinical variables
corr = vitalsigns_df[["Heart Rate", "Respiratory Rate", "Body Temperature", 
                      "Oxygen Saturation", "Systolic Blood Pressure", 
                      "Diastolic Blood Pressure", "Derived_BMI"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heattmap of Clinical Variables")
plt.show()
```

EXPLANATION
1. The correlation heatmap of clinical variables represent the relationship between different physiological measurements. It uses matrix where each cell's color and value shows the strength and direction of the correlation between two variables
2. The number within each cell represent the pearson correlation coefficient ranging from -1 to +1
3. The dark red(1.0) shows a perfect positive correlation, meaning the varibles increase or decrease together, it shows the correlation of variables with itself which is always perfect
4. The light colours, often blue with values close to 0, indicate a weak or no linear correlation between the variables, for example, the correlation between heart rate and respiratory rate is -0.01, showing a very weak negative correlation
5. The negative values in blue show negative correlation, where one variable tend to decrease as the other increases
6. The colour scale with a bar on the right side of the heat map gives a visual guide on how to interprete the correlation values, with  deep red indicating strong positive correlation and darker blue indicating stronger negative correlation, while lighter shades represent weaker correlations
